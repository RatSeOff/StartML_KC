```python
playstore = playstore_lower.copy().drop_duplicates('app').reset_index(drop=True)
```

```python
playstore['price'] =pd.to_numeric(playstore['price'].str.replace('$', ''))
```

```python
playstore[
    playstore['price']!=0
].head()
```

```python
df = playstore.copy()
```

```python
playstore.head()
```

```python
final_table = df.groupby(['category', 'type']
          ).agg(
    {'price':'mean', 'rating':'mean', 'reviews':'mean'}
).rename(
    columns={'price':'mean_price', 'rating':'mean_rating', 'reviews':'mean_reviews'}
)
```

```python
final_table.mean_price = final_table.mean_price.round(2)
final_table.mean_rating = final_table.mean_rating.round(1)
final_table.mean_reviews = final_table.mean_reviews.round(2)
final_table
```

```python
final_table.to_csv('final_table.csv', sep=',')
```
